```python
#доп функции:

def get_true_pdf_name(doc_data: dict) -> str:
    for index, file in enumerate(doc_data.get('Files', [])):
        if index == 0:
            return file.get('Id')
        
def get_pdf_name(path: Path) -> Path:
    try:
        for subpath in path.glob('*'):
            if subpath.name == 'document.json':
                doc_data = read_json(subpath)
        for subpath in path.glob('*'):
            if subpath.name == get_true_pdf_name(doc_data) + '.pdf':
                return subpath
    except TypeError:
        return
    
def search_pdf_by_path(path: Path) -> list:
    paths  = []
    if path.suffix == '.pdf':
        paths.append(path)
    else:
        for subpath in path.glob('*'):
            pdf = get_pdf_name(subpath)
            if pdf:
                paths.append(pdf)
    return paths
        
```

# создание  справочника сопоставлений. 

```python
appointed_senders = pd.DataFrame(df_predict[['sender', 'SENDER']].value_counts()).reset_index()
appointed_senders.rename(columns={0: 'count'}, inplace=True)
appointed_senders
appointed_senders = appointed_senders[appointed_senders['SENDER'] != 'не определён']
```

```python
appointed_senders[appointed_senders.sender =='аппарат правительства российской федерации']
```

```python
appointed_senders[appointed_senders.SENDER == 'евразийская экономическая комиссия']
```

```python
found_senders = set(appointed_senders.SENDER.to_list())
comparison = {}
for sender in found_senders:
    df_tt = appointed_senders[appointed_senders.SENDER == sender]
    if df_tt.shape[0] > 1:
        tt = df_tt.sort_values(by=['count'], ascending=False).iloc[0]
        if tt['count'] > 2:
            comparison[tt['SENDER']] = tt['sender']
    elif df_tt.shape[0] == 1:
        if df_tt.iloc[0]['count'] > 2:
            comparison[df_tt.iloc[0]['SENDER']] = df_tt.iloc[0]['sender']
```

```python
len(comparison), len(set(appointed_senders.SENDER.to_list())), len(appointed_senders)
```

```python
json_comparison = json.dumps(comparison, ensure_ascii=False)
with open('./data/comparison.json', 'w') as f:
    f.write(json_comparison)
```
